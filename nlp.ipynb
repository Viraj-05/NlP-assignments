```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    if text == "" or not isinstance(text, str):
        return [], ""
    
    text = text.lower()
    
    # remove urls
    text = re.sub(r'http\S+', '', text)
    
    # remove numbers
    text = re.sub(r'\d+', '', text)
    
    # remove repeated letters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    tokens = text.split()
    
    # remove small words except no, not
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]
    
    clean_sentence = " ".join(tokens)
    
    return tokens, clean_sentence
```

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
results = []

for text in sentences:
    tokens, clean = preprocess_text(text)
    
    print("Original:", text)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-"*40)
    
    results.append(tokens)
```

```python
for i, tokens in enumerate(results):
    if len(tokens) == 0:
        print("No tokens")
        continue
    
    total = len(tokens)
    unique = len(set(tokens))
    avg = sum(len(t) for t in tokens) / total
    
    print(f"Sentence {i+1}")
    print("Total:", total)
    print("Unique:", unique)
    print("Avg Length:", round(avg,2))
    print("-"*30)
```
